```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = list(metrics.keys())
train_acc = [round(metrics[model]['train_accuracy'], 4) for model in models]
test_acc = [round(metrics[model]['test_accuracy'], 4) for model in models]

fig, ax = plt.subplots(figsize=(12, 8))  # increase figure size

# Data
models = list(metrics.keys())
train_acc = [round(metrics[model]['train_accuracy'], 4) for model in models]
test_acc = [round(metrics[model]['test_accuracy'], 4) for model in models]

barWidth = 0.3
r1 = np.arange(len(train_acc))
r2 = [x + barWidth for x in r1]

plt.bar(r1, train_acc, color='b', width=barWidth, edgecolor='grey', label='train_acc')
plt.bar(r2, test_acc, color='r', width=barWidth, edgecolor='grey', label='test_acc')

# Adding xticks
plt.xlabel('Models', fontweight='bold' , fontsize=14)
plt.xticks([r + barWidth/2 for r in range(len(train_acc))], models, rotation=45, fontsize=14) # rotate x labels

# Add data values on top of the bars
for i in range(len(r1)):
    plt.text(x = r1[i]-0.1, y = train_acc[i]+0.01, s = train_acc[i], size = 14)
    plt.text(x = r2[i]-0.1, y = test_acc[i]+0.01, s = test_acc[i], size = 14)

plt.ylabel('Accuracy', fontsize=14)
plt.title('Model Performance', fontsize=14)
plt.legend()
plt.show()
```
